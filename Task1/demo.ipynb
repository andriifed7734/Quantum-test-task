#  Mountain Named Entity Recognition - Demo

This notebook demonstrates the inference capabilities of the fine-tuned DistilBERT model for identifying mountain names in unstructured text.

**Pipeline Overview:**
1. Load the fine-tuned Hugging Face model and tokenizer.
2. Process custom input texts using the custom `inference.py` script.
3. Visualize the extracted entities using `displaCy`.

```python
import warnings
from spacy import displacy
from inference import load_ner_pipeline, extract_mountains
from IPython.display import display, HTML

# Suppress transformer warnings for a clean output
warnings.filterwarnings('ignore')
```

## 1. Load Model Pipeline
Loading the `distilbert-base-uncased` model fine-tuned on our custom synthetic dataset.

```python
model_path = "./final_model" 
ner_pipe = load_ner_pipeline(model_path)
print("Pipeline loaded successfully!")
```

## 2. Define Test Cases
We will test the model on various scenarios:
- Standard sentences with well-known mountains.
- Sentences with multiple mountains.
- Negative examples (no mountains, but containing other proper nouns like cities).

```python
test_texts = [
    "Next year I want to climb Mount Everest and maybe K2.",
    "The capital of France is Paris, but the highest peak in the Alps is Mont Blanc.",
    "We traveled to New York, but there were no mountains there.",
    "Hoverla is the highest mountain in Ukraine, while Denali is the highest in North America.",
    "I bought a new car yesterday and drove it to the supermarket."
]
```

## 3. Run Inference and Visualize
We use `spacy.displacy` for a clean, visual representation of the recognized entities directly within the notebook.

```python
def highlight_mountains(text, extracted_entities):
    """
    Formats the extracted entities by changing the font color to light green 
    without applying any background highlights, and renders them via HTML.
    """
    # Sort entities in reverse order to prevent index shifting during string modification
    sorted_ents = sorted(extracted_entities, key=lambda x: x["start"], reverse=True)
    
    html_text = text
    for ent in sorted_ents:
        start = ent["start"]
        end = ent["end"]
        
        # Extract the exact substring representing the mountain
        mountain_name = html_text[start:end]
        
        # Wrap the name in an HTML span tag with a specific text color
        highlighted_name = f'<span style="color: #a6e22e; font-weight: bold;">{mountain_name}</span>'
        
        # Replace the original text with styled HTML span
        html_text = html_text[:start] + highlighted_name + html_text[end:]
        
    display(HTML(html_text))

# Run inference and display formatted results
for text in test_texts:
    mountains = extract_mountains(text, ner_pipe)
    highlight_mountains(text, mountains)
```
